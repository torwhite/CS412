```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial, exp

# Define the function
def f(x, lmbda):
    if x < 0:
        return 0
    else:
        return (exp(-lmbda) * lmbda**x) / factorial(x)

# Values of lambda
lambdas = [0.5, 1.0, 10.0]

# Generate x values
x_values = np.arange(0, 10, 1)

# Create subplots for each lambda
plt.figure(figsize=(12, 6))
for lmbda in lambdas:
    y_values = [f(x, lmbda) for x in x_values]
    plt.plot(x_values, y_values, label=f' = {lmbda}')

plt.xlabel('x')
plt.ylabel('f(x; )')
plt.title('Probability Mass Function for Poisson Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial, exp, log

# Define the logarithm of the function
def log_f(x, lmbda):
    if x < 0:
        return float('-inf')  # Log of 0 is negative infinity
    else:
        return -lmbda + x * log(lmbda) - log(factorial(x))

# Values of lambda
lambdas = [0.5, 1.0, .25]

# Generate x values
x_values = np.arange(0, 10, 1)

# Create subplots for each lambda
plt.figure(figsize=(12, 6))
for lmbda in lambdas:
    log_y_values = [log_f(x, lmbda) for x in x_values]
    plt.plot(x_values, log_y_values, label=f' = {lmbda}')

plt.xlabel('x')
plt.ylabel('log(f(x; ))')
plt.title('Logarithm of Probability Mass Function for Poisson Distribution')
plt.legend()
plt.grid(True)
plt.show()

```
